### Numpy is also called as Numerical Python used to perform various numerical and scientific computing in Python. Let us have a look at various mathematical functions, that are essential for data science, machine learning, engineering, and more.

```python
pip install numpy
```

Importing Numpy Package

```python
import numpy as np
```

Numpy is used to create various array operations like creating an NumPy array from a Python list or tuple.

```python
a = np.array([1, 2, 3, 4, 5])  # 1D array
b = np.array([(1, 2, 3), (4, 5, 6)])  # 2D array
print(a)
print(b)
```

Numpy can be used to initialize arrays with specific values like zeros and ones or any other matrices for linear algebra operations.

```python
zero_matrix = np.zeros((2, 3))  
print(zero_matrix)
```

```python
ones_matrix = np.ones((2, 2), dtype=int)
print (ones_matrix)
```

We can create sequences of numbers using Numpy for numerical computing.

```python
# Syntax : np.arange(start, stop, step size)

range = np.arange(1, 11, 2) 
print (range)
```

NumPy's product function is used to perform Matrix multiplication for 2D arrays.

```python
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6], [7, 8]])
dot_product = np.dot(c, d)
print(dot_product)
```

NumPy's built-in statistical functions allow for easy computation of key statistical measures, facilitating robust data analysis and interpretation.

```python
stats = np.array([1, 2, 3, 4, 5])
mean_val = np.mean(stats)
median_val = np.median(stats)
std_val = np.std(stats)
print(mean_val)
print(median_val)
print(std_val)
```

NumPy‚Äôs random submodule offers a wide range of functions to generate random numbers from various distributions, which is used to create stochastic models and performing Monte Carlo simulations.

```python
n = np.random.rand(2, 3)
print(n)
```

```python
import pandas as pd
```

Step 2: Loading the dataset and reading it using Pandas

Download the Red wine and White wine data from UCI Machine Learning Repository

```python
red_wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
white_wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'
```

Convert the downloaded dataset into CSV

```python
red_wine = pd.read_csv(red_wine_url, sep=';')
white_wine = pd.read_csv(white_wine_url, sep=';')
```

Let us create a column to distinguish between red and white wines

```python
red_wine['type'] = 'red'
white_wine['type'] = 'white'
```

Let us combine the csv's and make as one large dataset

```python
wine_data = pd.concat([red_wine, white_wine], ignore_index=True)
```

Let us Convert the dataset into a dataframe df

```python
df= wine_data
```

### Matplotlib is a comprehensive Python library for creating static, animated, and interactive visualizations. In this notebook we will be analyzing the Wine Dataset and create visualizations with Matplotlib.

```python
pip install matplotlib
```

Let's plot a Line graph to see how the average quality of wine varies with different levels of alcohol content.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
avg_quality = df.groupby('alcohol')['quality'].mean().reset_index()
plt.plot(avg_quality['alcohol'], avg_quality['quality'])
```

Let us style the plots by adding labels, titles to the plot to visualize it better and understand it better

```python
avg_quality = df.groupby('alcohol')['quality'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(avg_quality['alcohol'], avg_quality['quality'])
plt.title('Line Plot of Average Quality by Alcohol Content')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()
```

Let us create a Scatter Plot to visualize the chemical properties like volatile acidity and citric acid in wine samples 

```python
plt.scatter(df['volatile acidity'], df['citric acid'], c=df['quality'])
plt.xlabel('Volatile Acidity')
plt.ylabel('Citric Acid')
plt.colorbar(label='Quality')
plt.title('Scatter Plot of Volatile Acidity vs Citric Acid')
plt.show()
```

Let us create a Bar graph to visualize how the average volatile acidity varies across different wine quality ratings. 

```python
avg_volatile_acidity = df.groupby('quality')['volatile acidity'].mean().reset_index()

# Define a list of colors
colors = plt.cm.inferno(np.linspace(0, 1, len(avg_volatile_acidity)))

# Create a bar plot with different colors for each bar
plt.figure(figsize=(6, 6))
plt.bar(avg_volatile_acidity['quality'], avg_volatile_acidity['volatile acidity'], color=colors)
plt.xlabel('Quality')
plt.ylabel('Average Volatile Acidity')
plt.title('Bar Plot of Average Volatile Acidity vs. Quality')
plt.show()
```

Let us create a 3D scatter plot for red wine showing its alcohol content, density, and quality.

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 3D scatter plot for red wine
scatter = ax.scatter(red_wine['alcohol'], 
                     red_wine['density'], 
                     red_wine['quality'], 
                     c=red_wine['quality'],  # Color by quality
                     cmap='plasma',  # Choose a colormap
                     alpha=0.6)

ax.set_title('3D Scatter Plot of Red Wine: Alcohol Content, Density, and Quality')
ax.set_xlabel('Alcohol')
ax.set_ylabel('Density')
ax.set_zlabel('Quality')

# Add a color bar
clbar = plt.colorbar(scatter)
clbar.set_label('Quality')

# Rotate the plot for better visibility
ax.view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()
```
